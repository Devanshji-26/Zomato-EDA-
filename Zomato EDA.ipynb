```python
# Group by 'Aggregate rating' and 'Cuisines', and calculate the count
food_items = final_df[['Aggregate rating', 'Cuisines']].groupby(['Aggregate rating', 'Cuisines']).size().reset_index(name='Count')

# Sort the dataframe by 'Aggregate rating' in descending order
top_rated_cuisines = food_items.sort_values(by='Aggregate rating', ascending=False)

# Get the top 10 rated cuisines
top_10_cuisines = top_rated_cuisines.head(10)

# Display the result
print(top_10_cuisines)
```
